```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import seaborn as sns
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
%matplotlib notebook

rs = [8.88, 2.93, 2.13]

k = np.arange(0, 20, 0.1)
t = 1

ds = []
for r in rs:
    ds.append(np.exp(-r*t)*np.power(r*t, k)/factorial(k))

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(k, ds[0], label='Tricontinent w/ firmware Errors', color="#c94c4c")
ax.plot(k, ds[1], label='Tricontinent w/o firmware Errors', color="#eea29a", linestyle=":")
ax.plot(k, ds[2], label="Tecan", color="#b1cbbb")
ax.set(xlabel="Failures per 100 Runs", ylabel="Probability", title="Poisson Distribution for Failures per 100 Runs")
ax.legend()
plt.show()
```
