```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import seaborn as sns
sns.set(font_scale=1.5)
sns.set_style("whitegrid")
# %matplotlib notebook

rs = [10.07, 4.12, 4.35, 1]

k = np.arange(0, 20, 0.1)
t = 1

ds = []
for r in rs:
    ds.append(np.exp(-r*t)*np.power(r*t, k)/factorial(k))

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(k, ds[0], label='Tricontinent w/ firmware Errors', color="#c94c4c", linewidth=5)
ax.plot(k, ds[1], label='Tricontinent w/o firmware Errors', color="#eea29a", linestyle=":", linewidth=5)
ax.plot(k, ds[2], label="Tecan", color="#b1cbbb", linewidth=5)
ax.plot(k, ds[3], label="Ideal", color="#b1cbff", linewidth=5)
ax.set(xlabel="Failures per 100 Runs", ylabel="Probability", title="Poisson Distribution for Failures per 100 Runs")
ax.legend()
plt.show()
```
