```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("temp_delta.csv")
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.distplot(df['Delta'], label='Temp delta')
sns.distplot(df['Bottom Temp'], label='bottom temp')
sns.distplot(df['Top Temp'], label='top temp')
ax.axvline(1.6, label='plate temp')
ax.axvline(19.9, label='ambient temp', linestyle='--')
ax.set(xlabel='Temperature (C)', title='Temperatures from 96 Deep Well Plate')
ax.legend()
plt.show()
```
