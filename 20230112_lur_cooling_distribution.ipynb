```python
import os
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Import csv

```python
df = pd.read_csv("lur_cooling.csv")
# df['added'] = False
```

```python
df.head()
```

### Plotting

```python
cooling_times = []
cooling_times.append(list(df['LUR 1 (s)']))
cooling_times.append(list(df['LUR 2 (s)']))
cooling_times.append(list(df['LUR 3 (s)']))

fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(cooling_times, kde=True)
ax.set(title='Time to cool from 25°C to 5°C', xlabel=('Time (s)'))

label_list = []
label_list.append("LUR 3")
label_list.append("LUR 2")
label_list.append("LUR 1")
    
ax.legend(labels=label_list)
plt.show()
```
